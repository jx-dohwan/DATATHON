```python
파리바게뜨 = store['음식'].loc[store['음식']['상호명']=='파리바게뜨']
파리바게뜨
```

```python
파리바게뜨_위경도 = 파리바게뜨[['위도', '경도']]

import folium

m = folium.Map([36.3222756648811, 127.369163978367], zoom_start=14)
for 위도, 경도 in zip(파리바게뜨_위경도['위도'], 파리바게뜨_위경도['경도']):
  folium.Marker([위도, 경도]).add_to(m)

m
```
